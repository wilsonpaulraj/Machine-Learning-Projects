# **Sonar Rock vs Mine Prediction with Python**

This project aims to predict the object that the sonar is try to detect is whether a rock or a mine using machine learning in python.

[Sonar Dataset](https://www.kaggle.com/code/mahmudulhaqueshawon/sonar-rock-vs-mine-prediction)




Impoting Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Processing

```python
#loading the dataset to a pandas dataframe
sonar_data = pd.read_csv('/content/sonar data.csv', header=None)
sonar_data.head()
```

```python
sonar_data[60].value_counts()
```

```python
sonar_data.groupby(60).mean()
```

```python
#separate data and labels (input and output)
X = sonar_data.drop(columns=60, axis=1)
y = sonar_data[60]
```


splitting data into test and train data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)
```

Model Training

```python
model = LogisticRegression()

model.fit(X_train, y_train)
```

Model Evaluation

```python
#accuracy of the training data
training_data_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(training_data_prediction, y_train)
print(training_data_accuracy)
```

```python
#accuracy of the testing data
testing_data_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(testing_data_prediction, y_test)
print(testing_data_accuracy)
```
